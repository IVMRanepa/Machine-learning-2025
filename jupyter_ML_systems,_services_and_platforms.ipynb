**Задание 1:** Простая линейная регрессия

**Описание:** Реализуйте модель линейной регрессии для прогнозирования прибыли фудтрака на основе численности населения города.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Данные
population = np.array([1.1, 2.0, 3.0, 4.0, 5.0]).reshape(-1, 1)  # Численность населения (в миллионах)
profit = np.array([1.5, 2.0, 2.5, 3.5, 4.0])                   # Прибыль (в миллионах)

# Создание модели линейной регрессии
model = LinearRegression()
model.fit(population, profit)

# Прогнозирование
predicted_profit = model.predict(population)

# Визуализация
plt.scatter(population, profit, color='blue', label='Данные')
plt.plot(population, predicted_profit, color='red', label='Линейная регрессия')
plt.xlabel('Численность населения (млн)')
plt.ylabel('Прибыль (млн)')
plt.legend()
plt.title('Прогноз прибыли фудтрака')
plt.show()

# Вывод коэффициентов модели
print(f"Коэффициент наклона: {model.coef_[0]}")
print(f"Свободный член: {model.intercept_}")
```

**Выводы:**

Модель линейной регрессии успешно предсказывает прибыль на основе численности населения.

Коэффициенты модели показывают зависимость прибыли от численности.

**Задание 2:** Многомерная линейная регрессия

**Описание:** Предскажите цену дома на основе площади и количества комнат.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Данные: площадь (кв.м), количество комнат и цена (тыс. долларов)
X = np.array([[1200, 3], [1500, 4], [1700, 3], [2000, 5], [2500, 4]])
y = np.array([300, 400, 350, 500, 600])

# Создание модели многомерной линейной регрессии
model = LinearRegression()
model.fit(X, y)

# Прогнозирование цен для новых данных
new_houses = np.array([[1800, 4], [2200, 5]])
predicted_prices = model.predict(new_houses)

# Вывод результатов
print(f"Прогнозируемые цены для новых домов: {predicted_prices}")
```

**Выводы:**

Модель позволяет учитывать несколько факторов для прогнозирования цены.

Прогнозируемые цены соответствуют тренду в данных.

**Задание 3:** Логистическая регрессия для классификации

**Описание:** Определите вероятность поступления студента на основе результатов двух экзаменов.

```python
import numpy as np
from sklearn.linear_model import LogisticRegression

# Данные: результаты экзаменов и решение о поступлении (1 - поступил, 0 - не поступил)
X = np.array([[80, 85], [60, 70], [90, 95], [50, 55], [75, 80]])
y = np.array([1, 0, 1, 0, 1])

# Создание модели логистической регрессии
model = LogisticRegression()
model.fit(X, y)

# Прогнозирование вероятности поступления для нового студента
new_student = np.array([[70, 75]])
probability_of_admission = model.predict_proba(new_student)

print(f"Вероятность поступления нового студента: {probability_of_admission[0][1]:.2f}")
```

**Выводы:**

Логистическая регрессия позволяет оценить вероятность поступления.

Вероятность поступления зависит от результатов экзаменов.

**Задание 4:** K-Means кластеризация для анализа данных

**Описание:** Разделите данные о клиентах на группы по их покупательскому поведению.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Данные: покупки клиентов (сумма и количество покупок)
data = np.array([[100, 2], [200, 4], [300, 6], [400, 8], [500, 10]])

# Кластеризация с использованием K-Means (2 кластера)
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)

# Метки кластеров и центры кластеров
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Визуализация кластеров
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', label='Клиенты')
plt.scatter(centers[:, 0], centers[:, 1], color='red', marker='x', label='Центры кластеров')
plt.xlabel('Сумма покупок ($)')
plt.ylabel('Количество покупок')
plt.legend()
plt.title('Кластеризация клиентов')
plt.show()
```

**Выводы:**

Алгоритм K-Means успешно разделил клиентов на группы по их покупательскому поведению.

Центры кластеров показывают средние значения для каждой группы.

**Задание 5:** PCA для снижения размерности данных

**Описание:** Уменьшите размерность данных о лицах для визуализации.

```python
import numpy as np
from sklearn.decomposition import PCA

# Данные: искусственный набор признаков лиц (10 лиц с признаками в виде векторов размерности 5)
data = np.random.rand(10, 5)

# Снижение размерности до двух признаков с использованием PCA
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(data)

print("Сниженная размерность данных:")
print(reduced_data)
```

**Выводы:**

PCA позволяет уменьшить размерность данных без значительной потери информации.

Сниженные данные можно использовать для визуализации или дальнейшего анализа.
